## 0. Dice rolls

```python
import random
import statistics # Import the statistics module for the median() function

# 10 dice rolls using a list comprehension
rolls = [random.randint(1, 6) for _ in range(10)]
print(f"Original Rolls: {rolls}")
print("-" * 46)

# Sort the list in ascending order
ascending_rolls = sorted(rolls)
print(f"a) Ascending: {ascending_rolls}")

# Sort the list in descending order 
descending_rolls = sorted(rolls, reverse=True)
print(f"b) Descending: {descending_rolls}")

# Find the maximum and minimum value in the list 
max_value = max(rolls)
min_value = min(rolls)
print(f"c) Maximum Value: {max_value}, Minimum Value: {min_value}")

# Calculate the mean (average) of the list 
mean = sum(rolls) / len(rolls)
print(f"d) Mean (Average): {mean:.2f}")

# Find the median of the list using statistics.median()

median = statistics.median(rolls)
print(f"e) Median : {median}")

# Extract only the unique values from the list ---
unique_rolls = list(set(rolls))
print(f"f) Unique Values: {unique_rolls}")
```

```python
#a) Use list comprehension to create a list of squares from -10 to 10

# Create a list of numbers from -10 to 10
numbers = range(-10, 11)

# Square each number using a list comprehension
squares = [x**2 for x in numbers]

print(f"List of Squares: {squares}")

# Plot this list using matplotlib.

import matplotlib.pyplot as plt

# The x-axis values are the original numbers
x_values = list(numbers) # Convert range object to a list for clarity

# The y-axis values are the calculated squares
y_values = squares

# Create the plot
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b') # Plot the points with a line

# Add title and labels
plt.title('Squares of Numbers from -10 to 10')
plt.xlabel('Number (x)')
plt.ylabel('Square (x^2)')

# Show the plot
plt.show()
```

## 3. Chessboard

```python
# List comprehension for part

# Create a list named letters
letters = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']

# Create list Comprehension combination
combination = [letter + '1' for letter in letters] #For each letter, it performs string concatenation by appending the string '1'

print("a) Combination List:")
print(combination)
```

```python
# Define the letters (columns) and numbers (rows)
letters = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
# The ranks are 1 to 8. We'll use a string for easy concatenation.
numbers = ['1', '2', '3', '4', '5', '6', '7', '8']

# Initialize the empty 2D list (list of lists)
chessboard = []

# Loop through each rank (row) from 1 to 8
for number in numbers:
    # Initialize a list to hold the coordinates for the current row
    current_row = []

    # Loop through each file (column) from A to H
    for letter in letters:
        # Concatenate the file (letter) and the rank (number)
        coordinate = letter + number
        
        # Add the coordinate to the current row
        current_row.append(coordinate)
    
    # After completing the row, add it to the main chessboard list
    chessboard.append(current_row)

# Print the resulting chessboard list for verification
for row in chessboard:
    print(row)

# The full 2D list
print("\n b) Full 2D List:")
print(chessboard)
```

```python
# Define the file labels (columns) and rank labels (rows)
letters = ('A', 'B', 'C', 'D', 'E', 'F', 'G', 'H')
numbers = range(1, 9)

# Use a nested comprehension structure to build the coordinates
# The outer part iterates through each rank (1 through 8)
# The inner part creates the tuples for that specific rank, pairing each file with the current rank number.
chessboard_coordinates = tuple(
    # Inner: Create a tuple for a single rank, e.g., (('A', 1), ('B', 1), ..., ('H', 1))
    tuple((f, r) for f in letters) 
    # Outer: Loop through each rank
    for r in numbers
)

# Print the result to match the requested format
for combination_tuple in chessboard_coordinates:
    print(combination_tuple)
```

##### a. The probability that my original door has the rabbit is still 1/3. By switching, I am betting on the initial 2/3 probability that you were wrong, which is twice as likely as being right.

```python
import random

def monty_hall_simulation(num_simulations):
    """Simulates the Monty Hall problem for a given number of trials."""
    
    wins_stay = 0
    wins_switch = 0
    
    # 1. Doors are 1, 2, 3
    doors = [1, 2, 3]

    for _ in range(num_simulations):
        # 2. Randomly assign the door with the rabbit
        rabbit_door = random.choice(doors)
        
        # 3. Randomly pick an initial door
        initial_pick = random.choice(doors)

        # 4. Host opens a 'snake' door (must not be the rabbit door OR the player's initial pick)
        remaining_doors = [d for d in doors if d != rabbit_door and d != initial_pick]
        
        # In the common case (initial_pick is a snake), the host has only one choice.
        # If the initial_pick is the rabbit, the host can open either of the two remaining snake doors.
        if len(remaining_doors) == 2:
            host_open_door = random.choice(remaining_doors)
        else:
            host_open_door = remaining_doors[0]

        # 5. Determine the final remaining door
        final_remaining_door = [d for d in doors if d != initial_pick and d != host_open_door][0]

        # --- Tally the results ---

        # Strategy 1: STAY
        if initial_pick == rabbit_door:
            wins_stay += 1

        # Strategy 2: SWITCH
        # Switching means picking the final remaining door.
        if final_remaining_door == rabbit_door:
            wins_switch += 1

    # Calculate proportions
    proportion_stay = wins_stay / num_simulations
    proportion_switch = wins_switch / num_simulations

    return proportion_stay, proportion_switch

# List of simulation sizes to run
sim_sizes = [10, 100, 1000, 10000, 100000, 1000000]

results = {}
for size in sim_sizes:
    stay_p, switch_p = monty_hall_simulation(size)
    results[size] = (stay_p, switch_p)

# Print the final results
print("\n--- Simulation Results (Proportion of Wins) ---")
print("{:<10} | {:<10} | {:<10}".format("Simulations", "Stay", "Switch"))
print("-" * 35)
for size, (stay, switch) in results.items():
    print("{:<10} | {:<10.4f} | {:<10.4f}".format(size, stay, switch))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# a) Simulate 5000 uniformly distributed points and plot them

# Set the number of points for the simulation
N = 5000

# 1. Simulate N uniformly distributed points in the range (-1, 1)
# X and Y coordinates are independent and follow U(-1, 1)
X = np.random.uniform(-1, 1, N)
Y = np.random.uniform(-1, 1, N)

# 2. Calculate the Euclidean distance (d_i) of each point (x_i, y_i) from the origin (0, 0)
# d_i = sqrt(x_i^2 + y_i^2)
distances = np.sqrt(X**2 + Y**2)

# 3. Classify points as 'inner' (d_i < 1) or 'outer' (d_i > 1)
# The condition d_i < 1 defines the points inside the unit circle (radius 1)
inner_points_mask = distances < 1
outer_points_mask = distances > 1

# Extract the coordinates for plotting
X_inner = X[inner_points_mask]
Y_inner = Y[inner_points_mask]
X_outer = X[outer_points_mask]
Y_outer = Y[outer_points_mask]

# Plotting the results
plt.figure(figsize=(7, 7))

# Plot inner points with one color (e.g., blue)
plt.scatter(X_inner, Y_inner, color='blue', s=5, label='Inner Points ($d_i < 1$)')

# Plot outer points with another color (e.g., red)
plt.scatter(X_outer, Y_outer, color='red', s=5, label='Outer Points ($d_i > 1$)')

# Add the boundaries of the square and the circle for reference
# Square boundaries (from -1 to 1 on both axes)
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.gca().set_aspect('equal', adjustable='box') # Ensure the plot is a square
plt.grid(True, linestyle='--', alpha=0.6)

# Draw the unit circle for visualization (optional)
theta = np.linspace(0, 2*np.pi, 100)
circle_x = np.cos(theta)
circle_y = np.sin(theta)
plt.plot(circle_x, circle_y, color='green', linestyle='-', linewidth=2, label='Unit Circle ($d=1$)')

plt.title(f'Monte Carlo Simulation of {N} Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower right')
plt.show()

# b) Calculate the fraction and determine the convergence limit

# Count the number of inner and outer points
N_inner = len(X_inner)
N_outer = len(X_outer)

# Points exactly on the boundary (d_i = 1) are rare but must be accounted for
# N_boundary = N - N_inner - N_outer
# For the fraction calculation, we'll use the total N and the counts of inner/outer.

# Calculate the fraction between number of inner points and number of outer points
if N_outer > 0:
    fraction = N_inner / N_outer
else:
    fraction = np.inf # Should not happen with 5000 points

print("-" * 40)
print(f"Total simulated points (N): {N}")
print(f"Number of inner points (N_inner, d_i < 1): {N_inner}")
print(f"Number of outer points (N_outer, d_i > 1): {N_outer}")
print(f"Calculated fraction (N_inner / N_outer): {fraction:.4f}")
print("-" * 40)

# Theoretical Convergence
print("Convergence of the fraction (N_inner / N_outer) as N -> infinity:")
print("The fraction converges to the ratio of the areas: (Area of Circle) / (Area of Square excluding Circle)")
print(f"Theoretical Area Ratio = (pi * 1^2) / (2^2 - pi * 1^2) = pi / (4 - pi) ≈ {np.pi / (4 - np.pi):.4f}")
```
